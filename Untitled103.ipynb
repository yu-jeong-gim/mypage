```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# headers에 'User-Agent' 값 넣기
headers = {'User-Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"}

# url에 주소 넣기
url = "https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop"

# requests.get()으로 요청하기
r = requests.get(url, headers=headers)
# status_code 응답확인하기
r.status_code
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
print(soup)
```

```python
player_info = soup.find_all('tr', class_=['odd', 'even'])
```

```python
print(player_info[0])
print(len(player_info))
```

```python
number=[]
name=[]
position=[]
age=[]
nation=[]
team=[]
value=[]
```

```python
for info in player_info:
  player = info.find_all("td")
  #print(player)
  #player[0]
  #print(player[0])

  number.append(player[0].text)
  name.append(player[3].text)
  position.append(player[4].text)
  age.append(player[5].text)
  nation.append(player[6].img['alt'])
  team.append(player[7].img['alt'])
  value.append(player[8].text.strip())

  time.sleep(1)
  #print(value)
  #print(team)
  #print(nation)
  print(name)
  #print(number)
```

```python
df = pd.DataFrame(
        {"number":number,
         "name":name,
         "position":position,
         "age":age,
         "nation":nation,
         "team":team,
         "value":value}
    )
df
```

```python
df.to_csv('transfermarkt25.csv', index=False)
```
